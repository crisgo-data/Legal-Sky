```python
import pandas as pd
```

```python
df = pd.read_excel('prueba.xlsx')
```

```python
def dividir_por_comas(row):
    id_demanda = row['ID_DEMANDA']
    cod_reclamacion = str(row['COD_RECLAMACION'])
    valores_numericos = cod_reclamacion.split(',')
    return [(id_demanda, int(valor)) for valor in valores_numericos]
```

```python
lista_de_tuplas = df.apply(dividir_por_comas, axis=1).sum()
```

```python
nuevo_df = pd.DataFrame(lista_de_tuplas, columns=['ID_DEMANDA', 'COD_RECLAMACION'])
```
