```python
"""Vamos a sacara el dato de las aerolineas europeas por país en las cuales hay derecho a indemnización por retrasos"""
```

```python
url = 'https://www.aeropuertoinfo.com/info-sobre-vuelos/lista-de-aerolineas-en-europa/'
```

```python
html=req.get(url).text

soup=bs(html, 'html.parser')
```

```python
columns= soup.find_all('strong')
```

```python
nombre= []
for i in columns:
    nombre.append(i.text.strip())
print(nombre)
```

```python
columnas = nombre[1:3]
columnas
```

```python
aeros = soup.find_all('td')
```

```python
body = []
for i in aeros:
    body.append(i.text.strip())
print(body)

```

```python
body = body[3:]
body
```

```python
aerolineas = [body[0]]
for i, valor in enumerate(body[1:]):
    if (i+1) % 3 == 0:
        aerolineas.append(valor)

print(aerolineas)
```

```python

paises = [body[i] for i in range(1, len(body), 3)]
print(paises)
```

```python
import pandas as pd

# Lista de nombres de columna


# Crear un DataFrame vacío con las columnas especificadas
df = pd.DataFrame(columns=columnas)

print(df)
```

```python
df['Aerolínea'] = aerolineas
df['País'] = paises
```
